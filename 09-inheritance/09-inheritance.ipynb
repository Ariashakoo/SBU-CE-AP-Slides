  <p>
    کلمه کلیدی <strong> extends </strong> میگوید که شما کلاس جدیدی ایجاد کرده اید که از یک کلاس موجود ارث بری می کند
    <br>
    به کلاس موجود، <strong>ابرکلاس </strong>یا کلاس والد (پدر) گفته میشود و به کلاس جدید نیز، <strong>کلاس فرزند </strong>میگویند.
    در تصاویر، پیکان همیشه از فرزند به پدر رسم میشود :
  </p>


  <div class="images-big">
    <img src="img/inheritance5.png" alt="a sample of inheritance">
  </div>

  <p>
    بر اساس شکل بالا ، برنامه ای بنویسید که حقوق (salary) و پاداش (bonus) یک برنامه نویس را که از کارمند ارث بری میکند، چاپ کند.
  <br>
  </p>

</div>

```java
class Employee {
    float salary = 40000 ;
}

class Programmer extends Employee {
    int bonus = 10000 ;
    public static void main(String args[]){
        Programmer p = new Programmer();
        System.out.println("Programmer salary is : " + p.salary);
        System.out.println("Bonus of Programmer is : " + p.bonus);
    }
}
```

<div class="card" dir="rtl">
  <h3>مثالی از ارث بری چندگانه :</h3>
  <p>
    در مثال زیر، دو کلاس والد متدی به نام msg دارند و کامپایلر نمی داند کدام را برای فرزند ارث بری کند.
  </p>
  <p>
    لذا در هر حالت ( حتی عدم وجود متد های هم نام در کلاسهای والد) جاوا برای ارث بری چندگانه، خطای زمان کامپایل میگیرد.
  </p>

</div>




```java
class A{
    void msg(){
        System.out.println("hello");
    }
}
class B{
    void msg(){
        System.out.println("welcome");
    }
}
class C extends A,B { //suppose if it were
    Public Static void main(String args[]){
        C obj = new C();
        obj.msg(); //Now wich msg() method be invoked?
    }
}
```
